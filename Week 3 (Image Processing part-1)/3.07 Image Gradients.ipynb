##   <font style="color:rgb(134,19,348)">  What are Image Gradients  </font>

In very simple terms an image gradient is a directional change in the intensity or color of an image, The gradient of the image is one of the fundamental building blocks in image processing. They can also be thought of as derivatives in 1D.

Now I know it may not be clear to you now what gradients are, so ill further explain them by showing you some examples



- Look at point `1` and `2` they have constant color around them and so there is no change in intensity at point `1` & `2` and so we can say both these points have `0` Gradient.

- Now look at point `3` there is change in intensity in the x direction and it goes from black to white so we have high gradient in positive x direction.

- Similarly at point `4` there is change in intensity in the x direction but this time it goes from white to black so we have high gradient in negative x direction.

- At point `5` there is change in intensity in the y direction and it goes from black to white so we have high gradient in positive y direction.

- Similarly at point `6` there is change in intensity in the y direction but this time it goes from white to black so we have high gradient in negative y direction.

**Note:** Gradient in the y direction for point 3,4 is `0` and gradient in x direction for point 5,6 is `0`.

Now Its worth noting that I used the term `high gradient` because the reason is the change in direction is really high, its maximum as it goes from 0 to 255 or vice versa , in most cases the magnitude may not always be big but the gradient will still be there.

**Now gradients can obviously be in color too, see change in blue color gradient.** <br><br>



Now in color images, gradients are calculated for each channel separately. So at every individual pixel, we will have six numbers representing the gradient ( bx  ,  gx ,  rx ) & ( by  ,  gy ,  ry ) for B,G and R channels respectively. Its worth mentioning that in many applications we do not need color gradients , so in those cases the color images are converted to grayscale and then the gradients are calculated.

### <font style="color:rgb(134,19,348)">Note: So in general a Gradient has two properties, A magnitude which tells us how big the change in intensity is and the direction    </font>

Now to get the gradients final magnitude you will use the following formula: 

$$
G = \sqrt {G^2_x + G^2_y}
$$

Where $G_x$& $G_y$ are the magnitudes in x and y directions

To get the final Direction you will use this formula.

$$
\theta = \arctan \frac{G_y}{G_x}
$$
