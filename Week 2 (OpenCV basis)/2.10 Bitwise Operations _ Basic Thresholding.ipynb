###  <font style="color:rgb(134,19,348)"> The And Function</font>
And function takes two arrays of same shape and output 255 in all places where both images have 255 and 0 where one or neither have 255. Its just like the And logic gate.

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
````

````python
quater=cv2.imread('media/M2/rech.png',0)
half=cv2.imread('media/M2/recw.png',0)

finalimage = cv2.bitwise_and(quater,half)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(quater,cmap='gray');plt.title("Image 1"); 
plt.subplot(132);plt.imshow(half,cmap='gray');plt.title("Image 2"); 
plt.subplot(133);plt.imshow(finalimage,cmap='gray');plt.title("Result"); 
````

###  <font style="color:rgb(134,19,348)">  And Function with Mask</font>

In computer vision and image processing there a lot of functions which allow you to pass in an optional mask, now a mask is just binary image image of 255 and 0 i.e white and black. These masks tell the function on which which part of the image the function should run, if no mask is specified the function runs on the whole image if a mask is given then the function only looks at the part of the image where the mask is 255. The mask must have the same rows and cols as the image.

Now the And function also accepts a mask, we will first apply and function on two same images without mask and then with mask.

````python
white_block=cv2.imread('media/M2/andblock.png',0)
our_mask=cv2.imread('media/M2/andblockmask.png',0)

# if you pass 1 and 1 in AND function you get 1
finalimage = cv2.bitwise_and(white_block,white_block)  

# if you pass 1 and 1 in AND function you get 1 but only where the mask is 255
finalimage_withmask = cv2.bitwise_and(white_block,white_block, mask=our_mask)  

plt.figure(figsize=[20,20])
plt.subplot(141);plt.imshow(white_block,cmap='gray');plt.title("Image 1 and 2"); 
plt.subplot(142);plt.imshow(finalimage,cmap='gray');plt.title("Result without Mask"); 
plt.subplot(143);plt.imshow(our_mask,cmap='gray');plt.title("Mask"); 
plt.subplot(144);plt.imshow(finalimage_withmask,cmap='gray');plt.title("Result with Mask");
````

###  <font style="color:rgb(134,19,348)">  Not Function</font>

Like the popular Not gate the not function inverts the mask , for example 0 becomes 255 and 255 becomes 0

````python
our_mask=cv2.imread('media/M2/andblockmask.png',0)

mask_inv = cv2.bitwise_not(our_mask)  

plt.figure(figsize=[13,13])
plt.subplot(121);plt.imshow(our_mask,cmap='gray');plt.title("Normal Mask"); 
plt.subplot(122);plt.imshow(mask_inv,cmap='gray');plt.title("Inverted Mask"); 
````

###  <font style="color:rgb(134,19,348)">Thresholding</font>

There are times where we do not have transparent images so we do not have the alpha channel mask, in those cases we threshold the image to get the desired mask. There are different types of thresholding methods but at the moment we are going to discuss the most basic One. <br>

Here's how thresholding works, For every pixel, a threshold value is applied. If the pixel value is smaller than the threshold value, it is set to 0, otherwise it is set to a maximum value.

 [```cv2.threshold(src, thresh, max_value, type)```](https://docs.opencv.org/4.2.0/d7/d1b/group__imgproc__misc.html#gae8a4a146d1ca78c626a53577199e9c57) 

- `src`       input array multiple-channel (Your Image).
- `thresh`	  threshold value. (If you use THRESH_BINARY then all values above this are set to max_value.)
- `max_value` maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.
- `type`	  thresholding type. Most common one are THRESH_BINARY , THRESH_BINARY_INV


````python
import matplotlib.pyplot as plt
image=cv2.imread('media/M2/shapescontour.jpg',1)

# for getting a binary mask from threshold we must first convert the image to grayscale
image2gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret, th =  cv2.threshold(image2gray, 100, 255, cv2.THRESH_BINARY)

# THRESH_BINARY_INV Inverses the behaviour, now all pixels below this value are set to max_value
ret, th2 =  cv2.threshold(image2gray, 100, 255, cv2.THRESH_BINARY_INV)


plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image[:,:,::-1]);plt.title("Image")
plt.subplot(132);plt.imshow(th,cmap='gray');plt.title(" Thresholded Image");
plt.subplot(133);plt.imshow(th2,cmap='gray');plt.title("Inverse Thresholded Image");
````
