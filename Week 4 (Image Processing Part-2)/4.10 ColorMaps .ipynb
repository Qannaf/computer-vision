## <font style="color:rgb(134,19,348)">  ColorMaps </font> 

We have already seen color Maps in the OpenCV's first notebook where we met the default map of `Viridis` in matplotlib and we changed to `Grayscale` color map.




Now Linear data, or single intensity based data is often shown in interesting color maps, this is more useful, more appealing to the human eye and information is easily digested this way. A common example is how the temperature data of various cities is shown in color maps instead of boring single channel grayscale maps (where small changes are hardly  perceivable), in these color maps blue area represents the cooler areas and red represents the hotter areas.






## <font style="color:rgb(134,19,348)">  Applying Color Maps in OpenCV  </font> 
Now in OpenCV you can transform grayscale images to different ColorMaps 20 available color maps to be exact using `cv2.applyColorMap()` function. lets apply a color map to an image of Moon. you have to pass in the name of the color map you want to apply or you can just pass in index (from 0-19) of the map.

You can list of all available color maps and their codes [here](https://docs.opencv.org/4.1.0/d3/d50/group__imgproc__colormap.html#ga9a805d8262bcbe273f16be9ea2055a65)

```python
image = cv2.imread("media/M4/moon.jpg",0)

# lets apply the PLASMA color map, PLASMA is `4`
color = cv2.applyColorMap(image, cv2.COLORMAP_PLASMA)

plt.figure(figsize=[23,23])
plt.subplot(121);plt.imshow(image,cmap='gray');plt.title("Normal");plt.axis('off');
plt.subplot(122);plt.imshow(color[...,::-1]);plt.title("Color Map");plt.axis('off');
```

**Note: If you're still unsatisfied from these color maps then you even define your own custom color maps using the cv2.LUT() fucntion.**
